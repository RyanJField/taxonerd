```python
from taxonerd import TaxoNERD
from spacy import displacy
```

```python
nerd = TaxoNERD(prefer_gpu=False) # Set to True if you have a access to a GPU
```

```python
nerd.load("en_core_eco_weak_biobert")
```

## Looking for taxon names in text files

```python
doc = nerd.ner(open("./tests/test_data/test_txt/test1.txt", "r").read())
```

```python
displacy.render(doc, style="ent")
```

```python
doc = nerd.ner(open("./tests/test_data/test_txt/test2.txt", "r").read())
```

```python
displacy.render(doc, style="ent")
```
